```python
#항목 parsing 함수작성하기
def parse():
    try:
        ADDRESS = item.find("address").get_text()
        LATITUDE = item.find("latitude").get_text()
        LONGITUDE = item.find("longitude").get_text()
        TYPE = item.find("type").get_text()
        ZONEID = item.find("zoneId").get_text()
        ZONENAME = item.find("zoneName").get_text()

        return {
            "주소":ADDRESS,
            "위도":LATITUDE,
            "경도":LONGITUDE,
            "소유사":TYPE,#소유사 1는쏘카, 소유사2는그린카
            "차고지ID":ZONEID,
            "차고지명":ZONENAME
        }
    except AttributeError as e:
        return {
            "주소":None,
            "위도":None,
            "경도":None,
            "소유사":None,
            "차고지ID":None,
            "차고지명":None
        }
```
